# 1) Plot of the recursive Fibonacci execution time for n in [0,50]

```
# This is the fibonnaci optimized version, cuz the regular one is to slow for n > 35
def fibonacci_fast(n, fib_cache):

 if n == 0 or n == 1:
  return n
 if n in fib_cache:
  return fib_cache[n]
 else:
  fib_cache[n] = fibonacci_fast(n - 1, fib_cache) + fibonacci_fast(n - 2, fib_cache)
  return fib_cache[n]
```

```
from time import process_time

def time_recursive_fib(n):
  t0 = process_time()
  fib_cache = {}
  fibonacci_fast(n, fib_cache)
  tf =process_time()
  return tf-t0
time_recursive_fib(40)*100
```

```

import math
import matplotlib.pyplot as plt
timeList = [0]
n = 51
for i in range(1,n):
  timeList.append(time_recursive_fib(i))

plt.plot(range(1,n),timeList[1:])
plt.plot(range(1,n),range(1,n))
```

# 2) Plot of  the iterative Fibonacci linear and iterative Fibonacci logarithmic execution time n in [0,1000]

```
# fibI - Fibonacci Iterartive (forward) O(n lgn)
def fibI(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```
from time import process_time

def time_recursive_fib(n):
  t0 = process_time()
  fibI(n)
  tf =process_time()
  return tf-t0

time_recursive_fib(51)
```

```

import math
import matplotlib.pyplot as plt
timeList = [0]
n = 1001
for i in range(1,n):
  timeList.append(time_recursive_fib(i))

plt.plot(range(1,n),timeList[1:])
plt.plot(range(1,n),range(1,n))
```
